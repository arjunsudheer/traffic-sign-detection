```python
# import the YOLO model
from ultralytics import YOLO
# set the model to a medium-sized YOLOv8 model
model = YOLO("yolov8m.pt")
```

Dataset Link: https://universe.roboflow.com/vrobot8182-163-com/gtsdb-nzh3w/dataset/4

```python
# train the model with the dataset specific .yaml file
yaml_file_path = '/content/gdrive/MyDrive/Colab_Notebooks/Project_Engineering_Success/Object_Detection/GTSDB_Traffic_Signs_Object_Detection_Using_YOLOv8/GTSDB.v4i.yolov8/data.yaml'
model.train(data=yaml_file_path, epochs=100)
```

Video Link: https://youtu.be/QfnkMGgvFGs?feature=shared

```python
# use the trained model to predict the traffic signs in a video
video_file_path = '/content/gdrive/MyDrive/Colab_Notebooks/Project_Engineering_Success/Object_Detection/GTSDB_Traffic_Signs_Object_Detection_Using_YOLOv8/images/Driving_In_Germany.mp4'
results = model.predict(source=video_file_path, stream=True, save=True)
for i, r in enumerate(results):
    r.save(filename=f"/content/gdrive/MyDrive/Colab_Notebooks/Project_Engineering_Success/Object_Detection/GTSDB_Traffic_Signs_Object_Detection_Using_YOLOv8/images/results_{i}.jpg")
```
